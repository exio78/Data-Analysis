# **Men shoes sales, sizes (US) by month 2016**

Here I applied the same dummy variables and groupby but I also used pd.to_datetime top level function to filter out the months and pile them up in columns by a specifict year, in this case 2016.

```python
men = albundys.loc[albundys["Gender"] == "Male", ["Date", "Size (US)"]]
```

```python
men["Date"] = pd.to_datetime(men["Date"])
```

```python
men = men.loc[(men["Date"] >= "2016-01-01") & (men["Date"] <= "2016-12-31"), :] 
```

```python
men_month = pd.get_dummies(men["Date"].dt.month)
```

```python
men = pd.concat([men, men_month], axis = 1)
```

```python
men = men.groupby("Size (US)").sum(numeric_only = True)
```

**Calculating the mean for each sell of a shoe within a size during a month of the year for men**

```python
men["Mean"] = round(men.mean(axis = 1), 2)
```

```python
men
```

**Calculating the standard error**

the standard deviation for each mean of sizes divided by the root square of the number of observations, in this case we have 12

S / root square of N

```python
men["Standard Error"] = round(men.loc[: , 1:12].std(axis = 1) / np.sqrt(12), 2)
```

```python
men
```

**Margin of error**

We are going to use the T statistics table for the reliability factor with confidence intervals with unknown variance, the T statistics we are going to use is 2.20 because we have 12 - 1 and 0.5% of alpha 

```python
men["ME"] = round(men["Standard Error"] * 2.20, 2)
```

```python
men
```

**Confidence intervals**

```python
men["CI lower 95%"] = round(men["Mean"] - men["ME"], 2)
men["CI Higher 95%"] = round(men["Mean"] + men["ME"], 2)
```

```python
men
```

**Pairs per month**

```python
men["Number of pairs"] = round(men["CI Higher 95%"], 0)
```

```python
men
```

# Now let's compare two different Germany women shoes in the year 2016 to see if one outperfomrs the other in sales.

We are going to select two shops, GER1 and GER2, we are going to select only the year 2016, women shoes and the sells by sizes.

**Second shop**

```python
ger2 = albundys
```

```python
ger2["Year"] = ger2["Date"].dt.year
ger2["Month"] = ger2["Date"].dt.month
```

```python
ger2.loc[(ger2["Year"] == 2016) & 
         (ger2["Gender"] == "Female") & 
         (ger2["Shop"] == "GER2"), ["Size (US)", "Month"]]
```

```python
ger2 = ger2.loc[(ger2["Year"] == 2016) & 
         (ger2["Gender"] == "Female") & 
         (ger2["Shop"] == "GER2"), ["Size (US)", "Month"]]
```

```python
dummies2 = pd.get_dummies(ger2["Month"])
```

```python
ger2 = pd.concat([ger2, dummies2], axis = 1).drop("Month", axis = 1)
```

```python
ger2 = ger2.groupby("Size (US)").sum()
```

**Let's calculate the mean per sell size shoes for these two shops**

```python
ger1_mean = round(np.mean(ger1, axis = 1), 2)
ger2_mean = round(np.mean(ger2, axis = 1), 2)
```

**Now the sample variance for these two shops by sell size shoes**

```python
ger1_s_var = round(ger1.var(axis = 1),  2)
ger2_s_var = round(ger2.var(axis = 1),  2)
```

**Now we are going to calculate the pooled variance**

```python
pooled = ((12 - 1) * ger1_s_var + (12 - 1) * ger2_s_var) / (12 + 12 - 2)
```

**The margin of error of these two variables or shops**

The T statistics we are going to use is 12 + 12 - 2, alpha /2 degrees of freedom for a 95% confidence interval so alpha would be 0.025, and from the table of T statistic we get 2.07

```python
ME = round(2.07 * np.sqrt(pooled / 12 + pooled / 12), 2)
```

**Time to calculate the confidence intervals**

```python
mean_difference = ger1_mean - ger2_mean
```

```python
ci_low = mean_difference - ME
ci_high = mean_difference + ME
```

**Now let's build the dataframe** 

```python
insights = pd.concat([ger1_mean, ger2_mean, ger1_s_var, ger2_s_var, pooled, ME, ci_low, ci_high], axis = 1)
```

```python
insights = insights.rename(columns = {0 : "mean_ger1",
                          1 : "mean_ger2",
                          2 : "ger1_var",
                          3 : "ger2_var", 
                          4 : "pooled_var",
                          5 : "ME", 
                          6 : "ci_low", 
                          7 : "ci_high"})
```

```python
insights
```

# Now estimate the 90% confidence intervals and check what changes you see

It would be the same process as before but this time we are going to change the degrees of freedom, these grees are the sum of the two varibale lenght minus 2, it would be 22 on the y axis of the t students table and 0.05 on the x axis. The number is 1.72 

```python
ME_90 = round(1.72 * np.sqrt(pooled / 12 + pooled / 12), 2)
```

```python
ci_low_90 = mean_difference - ME_90
ci_high_90 = mean_difference + ME_90
```
