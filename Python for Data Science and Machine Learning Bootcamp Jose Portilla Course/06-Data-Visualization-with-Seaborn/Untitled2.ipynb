```python
import seaborn as sns

%matplotlib inline

tips = sns.load_dataset("tips")
flights = sns.load_dataset("flights")
tips.head()
```

```python
flights.head()
```

```python
tc = tips.corr()
tc
```

```python
sns.heatmap(tc, annot = True, cmap = "coolwarm", )
```

```python
fp = flights.pivot_table(index = "month", columns = "year", values = "passengers")
fp
```

```python
sns.heatmap(fp, cmap = "coolwarm", linewidth = 6, linecolor="white")
```

```python
sns.clustermap(fp, cmap = "coolwarm", standard_scale = 20)
```
