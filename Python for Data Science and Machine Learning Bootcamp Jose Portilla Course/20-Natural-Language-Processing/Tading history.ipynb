```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
trading_2020 = pd.read_excel("timeline trading 2020.xlsx")
```

```python
trading_2020.info()
```

```python
trading_2020.columns
```

```python
trading_2020_n = pd.DataFrame(data = trading_2020, index = trading_2020.columns, columns = ["result"])
```

```python
trading_2020_n["result"] = trading_2020.iloc[0]
```

```python
trading_2020_n
```

```python
trading_2020_index = pd.DataFrame(data = trading_2020, columns = ["result"])

trading_2020_index["result"] = trading_2020.iloc[0]
trading_2020_index
```

```python
figure = plt.figure(figsize=(50,15), facecolor = "#2f3640")

ax = plt.axes()
ax.set_facecolor("#2f3640")

plt.plot(trading_2020_n, marker = "o", markerfacecolor = "#EA2027", markersize = 20, ls = "-", lw = 10, color = "#0652DD")
plt.title("Trading history line 2020", fontsize = 50, color = "#dcdde1")
plt.xlabel("months", fontsize = 50, color = "#dcdde1")
plt.ylabel("results in %", fontsize = 50, color = "#dcdde1")
plt.tight_layout
```
