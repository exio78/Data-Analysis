```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x **2
```

```python
x
```

```python
y
```

```python
#functional method
```

```python
plt.plot(x, y)
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("Title")
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, "r")

plt.subplot(1, 2, 2)
plt.plot(y, x, "b")
```

```python
figure = plt.figure()

axes = figure.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)

axes.set_xlabel("X label")
axes.set_ylabel("y label")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.20, 0.50, 0.4, 0.3])

axes1.plot(x, y)
axes1.set_title("bigger")

axes2.plot(y, x)
axes2.set_title("smaller")
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2)

axes[0].plot(x, y)
axes[0].set_title("first plot")

axes[1].plot(y, x)
axes[1].set_title("Second Plot")
```

```python
axes
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 1, figsize=(8, 2))

axes[0].plot(x, y)
axes[1].plot(y, x)

plt.tight_layout()
```

```python
fig.savefig("my_picture.png", dpi = 200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x ** 2, label = "x squared")
ax.plot(x, x **3, label = "x cubed")

ax.legend(loc = 5)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color = "#6c5ce7", lw = 3, alpha = 0.5, ls ="-", marker = "o", markersize = 10, markerfacecolor = "#ffeaa7", 
       markeredgewidth = 2, markeredgecolor = "#00b894")
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color = "#6c5ce7", lw = 2, ls = "--")

ax.set_xlim([0, 1])
ax.set_ylim([0, 2])
```

```python
data = pd.read_csv("sample_data.csv")
data
```

```python
alpha = 2/5
beta = 100
predictions = alpha * data.square_meters + beta

fig,ax = plt.subplots()
ax.scatter(data.square_meters, data.price)
ax.plot(data.square_meters,predictions,c="tab:gray")
ax.vlines(data.square_meters, data.price, predictions,color='red')
ax.set(
    xlabel="square meters",
    ylabel="price"
)
```
