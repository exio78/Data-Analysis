# **What is the probability we earn K number of trades within N number of trades within a measure of time?  üìê**

In the stock market üìà as in any other field where we want to apply inferencial statistic and assume the most likely outcome where we have:

1. Independent events 
2. Two possible outcomes (the one we are looking for and the other one we are not looking for) 
3. Constant probability for each individual outcome inside of N number of trials
4. Finite or discrete sample space 

We can apply a Binomial Distribution, a kind of probability distribution of a random variable to define P(x) ‚õàÔ∏èüå§Ô∏èüåßÔ∏è

```python
import matplotlib.pyplot as plt 
plt.style.use("fivethirtyeight")

for number_trades in range(1, 13):

    def factorial(n): # define the function to get the factorial of any number
        
        if n == 0:
            return 1 # the factorial of 0 is 1 so we need to return 1 if we place 0
        else:
            return factorial(n - 1) * n

    K = 0 # exact number of items we want to succeed for N = number_trades trials
    N = number_trades # number of trials we are going to make
    Iterate = number_trades + 1 # the last number of the iteration is exclusive
    N_minus_K = N - K # We are going to use this in the combination formula

    fact_N = factorial(N) 

    binomial_dist = [] # binomial distribution y axis, probability of k in N trials
    x_axis = [] # number of trades won in N trials x axis 

    for i in range(Iterate):

        #Combination formula where N chooses K
        Combination = fact_N / (factorial(N_minus_K) * factorial(K)) 
        #print(f"combination {Combination}")

        # Probability of earning a trade to the K chooses of N trials
        P_earn_trade = (1/2) ** K

        # Probability of not earning a trade to the  N minus K  
        P_lose_trade = (1 - 1/2) ** (N - K)

        PDBR = round(Combination * P_earn_trade * P_lose_trade, 4) #Getting the product

        K += 1 # increase the exact number of items we want to suceed 
        N_minus_K -= 1 # We need to update this variable, altering k does not alter this

        # storing the values for the x and y axis
        binomial_dist.append(PDBR)
        probability_x = K - 1 
        x_axis.append(probability_x)

        #print()
        #print(f"P({K - 1}) Binomial: {PDBR}")
        #print()

    #print(binomial_dist)
    #print(x_axis)

    # Visualize in a barchart the Binomial distribution of the trades in the Stock Market

    plt.bar(x_axis, binomial_dist, color = "#a29bfe")
    plt.title(f"Binomial Distribution {K - 1} Trials")
    plt.xlabel(f"Earn a trade for {N} trials")
    plt.ylabel("Probability of getting a number of earned trades")
    plt.xticks(x_axis)
    plt.subplots_adjust(bottom = 0.08, left = 0.10)
    plt.show()   
```
